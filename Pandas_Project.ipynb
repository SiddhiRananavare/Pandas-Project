```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [24, 27, 22, 32, 29],
    'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
    'Salary': [45000, 54000, 50000, 62000, 47000]
}

df = pd.DataFrame(data)
```

```python
#summary statistics of age and salary
df[['Age', 'Salary']].describe()
```

```python
hr_avg_salary = df[df['Department'] == 'HR']['Salary'].mean()
print("Average HR Salary:", hr_avg_salary)

```

```python
#  Add a new column 'Bonus' = 10% of Salary
df['Bonus'] = df['Salary'] * 0.10
print("DataFrame after adding Bonus column:")
print(df)
```

```python
#  Filter employees aged between 25 and 30
age_filter = df[(df['Age'] >= 25) & (df['Age'] <= 30)]
print("\nEmployees aged between 25 and 30:")
print(age_filter)
```

```python
# Group by Department and calculate average salary
dept_avg_salary = df.groupby('Department')['Salary'].mean()
print("\nAverage salary by department:")
print(dept_avg_salary)
```

```python
#Sort DataFrame by 'Salary' in ascending order and save to CSV
sorted_df = df.sort_values(by='Salary', ascending=True)
sorted_df.to_csv("sorted_employees.csv", index=False)
```
